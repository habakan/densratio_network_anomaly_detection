```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def entropy(data, window_s):
  
  def calc(hensu):
    hensu = hensu.value_counts()
    num = []
    for i in hensu:
      num.append(i/window_s * np.log2(i/window_s))
    goukei = sum(num) * -1
    return goukei

  entro = []
  for x in range(0, len(data), window_s):
    window = data.iloc[x:x+window_s, :]
    hensu = np.array(window.apply(lambda y: calc(y)))
    entro.append(hensu)
  
  return np.array(entro).T
```

```python
data2 = pd.read_csv("./pcap_data/week2_mon.csv")
data3 = pd.read_csv("./pcap_data/week2_thur.csv")
```

```python
data2_e = sum(entropy(data2, 100))
```

```python
data3_e = sum(entropy(data3, 100))
```
